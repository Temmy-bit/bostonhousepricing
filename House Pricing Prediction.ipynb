### Boston House Pricing Dataset

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
boston_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([boston_df.values[::2, :], boston_df.values[1::2, :2]])
target = boston_df.values[1::2, 2]
```

```python
X = dataframe.iloc[:,:-1]
y = dataframe.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=26)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_train.columns
```

```python
X_train.iloc[13]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

```python
import pickle

# with open('scaler.pkl', 'wb') as f:
#     pickle.dump(sc, f, protocol=pickle.HIGHEST_PROTOCOL)
pickle.dump(sc, open('scaler.pkl', 'wb'))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
print(lr.intercept_)
```

```python
lr.get_params()
```

```python
lr.score(X_test,y_test)
```

```python
reg_pred = lr.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(reg_pred,y_test)
plt.show();
```

```python
residuals = y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
plt.show();
```

```python
# scatter plot of pred with respect to residuals
plt.scatter(reg_pred,residuals)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,reg_pred),mean_squared_error(y_test,reg_pred),np.sqrt(mean_squared_error(y_test,reg_pred))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,reg_pred)
score
```

```python
1-(1-score)*(len(y_test)-1) / (len(y_test)-X_test.shape[1]-1)
```
